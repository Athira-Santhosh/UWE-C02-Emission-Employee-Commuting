```python
import matplotlib.pyplot as plt

# Data
car_types = ['Petrol', 'Diesel', 'Plug in Hybrid', 'Electric', 'Hybrid (50/50 average)']
emissions_from_fuel = [0.26379, 0.27332, 0.10601, 0, 0]
emissions_from_electricity = [0, 0, 0.04152, 0.08116, 0]
well_to_tank_fuel = [0.07325, 0.0667, 0.03927, 0.01955, 0]
transmission_distribution_electricity = [0, 0, 0.00359, 0.00703, 0.001794493]
total_emissions_per_mile = [0.33704, 0.34002, 0.19039, 0.10774, 0.00179]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting each component separately
ax.bar(car_types, emissions_from_fuel, label='Emissions from Fuel (Kg CO2e)', color='skyblue')
ax.bar(car_types, emissions_from_electricity, label='Emissions from Electricity (Kg CO2e)', bottom=emissions_from_fuel, color='salmon')
ax.bar(car_types, well_to_tank_fuel, label='Well To Tank (Fuel)', bottom=[i+j for i,j in zip(emissions_from_fuel, emissions_from_electricity)], color='lightgreen')
ax.bar(car_types, transmission_distribution_electricity, label='Transmission & Distribution (Electricity)', bottom=[i+j+k for i,j,k in zip(emissions_from_fuel, emissions_from_electricity, well_to_tank_fuel)], color='orange')
ax.bar(car_types, total_emissions_per_mile, label='Total emissions (Kg CO2e per mile)', bottom=[i+j+k+l for i,j,k,l in zip(emissions_from_fuel, emissions_from_electricity, well_to_tank_fuel, transmission_distribution_electricity)], color='purple')

ax.set_ylabel('Emissions (Kg CO2e)')
ax.set_title('Emissions Conversion Factors for Different Types of Cars')
ax.legend()

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Petrol and Diesel Cars: Petrol cars emit approximately 0.264 kg CO2e per mile, while diesel cars emit slightly more at around 0.273 kg CO2e per mile. Both types of cars contribute to emissions from fuel combustion, with petrol having a slightly higher emission factor.

Plug-in Hybrid Cars: These cars have a lower emission factor compared to traditional petrol and diesel cars. The emissions from fuel combustion for plug-in hybrid cars are approximately 0.106 kg CO2e per mile. Additionally, there are emissions associated with the transmission and distribution of electricity, which adds up to about 0.190 kg CO2e per mile.

Electric Cars: Electric cars have zero emissions from fuel combustion since they run solely on electricity. However, there are emissions associated with the transmission and distribution of electricity, which contribute to the total emissions. Electric cars emit about 0.108 kg CO2e per mile, with the majority of emissions coming from the transmission and distribution of electricity.

Hybrid (50/50 average): This category likely refers to hybrid cars that have a mix of petrol or diesel engine and electric motor. The emission factors provided for this type of car are relatively low, indicating lower emissions compared to traditional petrol and diesel cars. The emission factor for transmission and distribution of electricity is negligible.

```python
import matplotlib.pyplot as plt

# Data
activities = ['Office Equipment', 'Heating', 'Homeworking (Office Equipment + Heating)']
emissions = [0.03144, 0.30234, 0.33378]

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(activities, emissions, color='skyblue')
plt.xlabel('Activity')
plt.ylabel('Emissions (kg CO2e)')
plt.title('Emissions from Working from Home')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Homeworking (Office Equipment + Heating) contributes the most to emissions among the listed activities. This indicates that the combined emissions from using office equipment and heating while working from home have a significant impact on the environment.

Heating contributes a substantial amount to emissions, suggesting that energy consumption for heating during remote work plays a significant role in overall emissions.

Office Equipment contributes the least to emissions among the listed activities, but it still adds to the overall carbon footprint. This includes the energy consumption of devices such as computers, monitors, printers, etc., used for work purposes at home.
